#### Cryosat Level-1b SARIn data visualization.


```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import netCDF4 as nc
import numpy as np
from pyrsimg import readTiff, imgShow, second_to_dyr
```

```python
## cryosat-2 sarin data.
path_cryo2_tempo_eolis = 'data/cryosat-2/CS_OFFL_THEM_POINT_SOUASIAEA_2021_08_+95424706_+29318642_V202.nc'  # GOPM:LRM-GOP
```

### Data reading

```python
## 1. cryosat-2 SARIn data
cryo2_tempo_eolis = nc.Dataset(path_cryo2_tempo_eolis)
```

```python
lon = cryo2_tempo_eolis['lon'][:]
lat = cryo2_tempo_eolis['lat'][:]
time = cryo2_tempo_eolis['time'][:]
print(time.shape)
```

### Visualization from global perspective

```python
proj = ccrs.PlateCarree(central_longitude=0.0)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection=proj)
# ax.set_extent([95.5, 97, 29, 30.4])
ax.stock_img()
ax.plot(cryo2_tempo_eolis['lon'][:], cryo2_tempo_eolis['lat'][:], marker='.', markersize=1, linewidth=0.0, color='red')
ax.gridlines(draw_labels=True, linewidth=1, color='gray', linestyle='--')
ax.legend()
```
